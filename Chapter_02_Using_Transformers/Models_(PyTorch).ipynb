BERT does not really "work" intelligently yet.

What happens is:

1. Input sentence
text = "I love NLP"
2. Tokenization
["[CLS]", "i", "love", "nl", "##p", "[SEP]"]

Each token gets converted to an ID:

[101, 1045, 2293, 17953, 2361, 102]
3. Embedding Layer

BERT converts each token ID into a vector of size 768.

Example:

1045 ("i")
↓
[0.23, -1.1, 0.78, ...]
4. Transformer Encoder Layers

The vectors pass through 12 encoder layers.

Each layer uses self-attention.

For example:

I love NLP

When processing NLP, BERT looks at:

I
love
NLP

and decides which words are important.

Maybe:

NLP attends 70% to love
NLP attends 20% to I
NLP attends 10% to itself

This is self-attention.

5. Output

BERT produces contextual embeddings.

Shape:

(batch_size, sequence_length, hidden_size)

For one sentence:

(1, 6, 768)

Each token now has a 768-dimensional vector containing information about the entire sentence.

Why doesn't your model know English?

Because:

model = BertModel(config)

creates random weights.

Imagine:

import torch

torch.randn(...)

The attention matrices are random.

The embeddings are random.

The feed-forward layers are random.

So the model architecture exists, but it has no knowledge.
